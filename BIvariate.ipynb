## T-Test

## Independent Sample
## Different group(male, female) but same condition(salary)

## (is significantly same the salary of male and female candidate?)

```python
from scipy.stats import ttest_ind

preprocessedPlacement = preprocessedPlacement.dropna()

male = preprocessedPlacement[preprocessedPlacement["gender"] == "M"]["salary"]

female = preprocessedPlacement[preprocessedPlacement["gender"] == "F"]["salary"]

ttest_ind(male, female)
```

## Hypothesis testing

```python
test_ans= ttest_ind(male, female)

test_ans.pvalue
```

```python

print("Null Hypothesis : There is no significantly same the salary of male and female candidate")
    
print("Alternative Hypothesis : There is significantly same the salary of male and female candidate")

messageList = []

Hypo_pvalue = test_ans.pvalue

if(Hypo_pvalue < 0.05):
    
    print("Reject Null Hypothesis")
    print("Accept Alternate Hypothesis")
    
    messageTrueHypothesis = "Accept Alternate Hypothesis"
            
else:
    
    print("Reject Alternate Hypothesis")
    
    print("Accept Null Hypothesis")

    messageTrueHypothesis = "Accept Null Hypothesis"

messageList.append(messageTrueHypothesis)
```

```python
messageList
```

## Dependent Sample
## Same group(male) but different condition(ssc_p, hsc_p)

```python

from scipy.stats import ttest_rel

male = preprocessedPlacement[preprocessedPlacement["gender"] == "M"]["ssc_p"]

male_1 = preprocessedPlacement[preprocessedPlacement["gender"] == "M"]["hsc_p"]

ttest_rel(male, male_1)
```

```python
t_ans = ttest_rel(male, male_1)

t_ans.pvalue
```

```python

print("Null Hypothesis : There is no significantly same ssc_p and hsc_p of male candidate")
    
print("Alternative Hypothesis : There is significantly same ssc_p and hsc_p of male candidate")

messageHypothesis = []

testT = t_ans.pvalue

if(testT < 0.05):
    
    print("Reject Null Hypothesis")

    print("Accpet alternative Hypothesis")
    
    messageBox = "Accpet alternative Hypothesis"
    
else:
    
    print("Reject Alternative Hypothesis")
    
    print("Accept Null Hypothesis")
    
    messageBox = "Accept Null Hypothesis"
    
messageHypothesis.append(messageBox)
```
